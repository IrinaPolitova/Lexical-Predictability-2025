```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def get_stats(driver, wait, query):
    """
    Переходит на страницу поиска, вводит запрос, нажимает "Искать" и извлекает статистику.
    Возвращает кортеж: (texts_count, examples_count).
    Если статистика не найдена – возвращает (None, None).
    """
    # Переходим на страницу поиска
    driver.get("https://ruscorpora.ru/search")
    
    try:
        # Переключаемся на вкладку "Поиск точных форм" (если требуется)
        try:
            exact_tab_btn = wait.until(
                EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Поиск точных форм')]"))
            )
            exact_tab_btn.click()
        except Exception:
            # Если кнопка не найдена, возможно, вкладка уже активна
            pass

        # Ждем, пока появится поле ввода внутри панели поиска
        search_input = wait.until(
            EC.visibility_of_element_located((
                By.CSS_SELECTOR,
                "#exact-search-panel .exact-form-search-fields .the-input__field input.the-input__input"
            ))
        )
        search_input.clear()
        search_input.send_keys(query)
        
        # Нажимаем кнопку "Искать"
        search_button = wait.until(
            EC.element_to_be_clickable((
                By.CSS_SELECTOR,
                "#exact-search-panel button.corpus-search-tab__exact-search"
            ))
        )
        search_button.click()

        # Ждем появления блока статистики в разделе "Запрос"
        stats_container = wait.until(
            EC.visibility_of_element_located((
                By.CSS_SELECTOR,
                ".the-header--query .the-header__stats-counter"
            ))
        )
        # Находим все элементы <li> внутри списка статистики
        stats_items = stats_container.find_elements(By.TAG_NAME, "li")
        if len(stats_items) >= 2:
            texts_count = stats_items[0].text  # например, "60 текстов"
            examples_count = stats_items[1].text  # например, "63 примера"
        else:
            texts_count, examples_count = None, None

    except Exception as e:
        print(f"Ошибка при поиске для запроса '{query}': {e}")
        texts_count, examples_count = None, None

    return texts_count, examples_count

# Список поисковых запросов
queries = words

# Инициализируем драйвер и WebDriverWait
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 20)

results = []

for query in queries:
    texts, examples = get_stats(driver, wait, query)
    results.append({"query": query, "texts": texts, "examples": examples})
    # Добавляем небольшую задержку между запросами
    time.sleep(2)

driver.quit()

# Преобразуем результаты в pandas DataFrame и выводим
df = pd.DataFrame(results)
print(df)
```

```python
def add_verbs(tags):
    if 'PRTF' in tags or 'GRND' in tags or 'INFN' in tags:
        tags.append('VERB')
    return tags
```

```python
def add_column_with_poses_1(row):
    word = row['Слово']
    id = row['Пропуск']
    sem = row['Семантическое поле']
    try:
        sem = sem.lower()
        if sem.endswith(' '):
            sem = sem[:len(sem)-1]
            print(sem)
        if sem.startswith(' '):
            sem = sem[1:]
            print(sem)
    except:
        sem = sem
    
    words_analized = [morph.parse(token) for token in word.split()]
    
    parses = []
    for a in words_analized:
        parses.extend(a)
    
    tags = []
    for parse in parses:
        tags.append(parse.tag.POS)
    tags = add_verbs(tags)

    true_pos = dict_of_poses[dict_of_true_values_1[id][0]]

    if true_pos in tags:
        correct_pos = 1
    else:
        correct_pos = 0

    true_sem = dict_of_true_values_1[id][1]

    if sem == true_sem:
        correct_sem = 1
    else:
        correct_sem = 0

    row['correct_pos'] = correct_pos
    row['correct_sem'] = correct_sem
    
    return row

sem_1 = sem_1.apply(add_column_with_poses_1, axis=1)
sem_1.to_csv('sem_1.csv', index=False)
```

```python
from itertools import groupby

spl = '\n'

ud_data_splitted = []
for k, g in groupby(ud_data, lambda x: x == spl):
    if not k:
        a = list(g)
        new_a = []
        for el in a:
            if el.startswith('#'):
                pass
            else:
                el = el.split('\t')
                if el[3] != 'PUNCT' and el[3] != '_':
                    new_a.append(el)
        ud_data_splitted.append(new_a)
```

```python
def get_poses(words):
    poses = []
    for word in words:
        poses.append(word[3])
    return poses

def get_pentagrams(words):
    poses = get_poses(words)
    pentagrams  = []
    if len(poses) >= 5:
        for i in range(len(poses)-4):
            new_penta = poses[i:i+5]
            pentagrams.append(new_penta)
    return pentagrams

def get_trigrams(words):
    poses = get_poses(words)
    trigrams  = []
    if len(poses) >= 3:
        for i in range(len(poses)-2):
            new_tri = poses[i:i+3]
            trigrams.append(new_tri)
    return trigrams
```

```python
all_pentas = []
for sentence in ud_data_splitted:
    new_pentas = get_pentagrams(sentence)
    all_pentas.extend(new_pentas)
```

```python
all_tris = []
for sentence in ud_data_splitted:
    new_tris = get_trigrams(sentence)
    all_tris.extend(new_tris)
```

```python
from collections import Counter
cnt_pentas = Counter()
cnt_tris = Counter()

for poses in all_pentas:
    poses = tuple(poses)
    cnt_pentas[poses] += 1

for poses in all_tris:
    poses = tuple(poses)
    cnt_tris[poses] += 1
```

```python
all_tags = []

for tri in all_tris:
    for tag in tri:
        if tag not in all_tags:
            all_tags.append(tag)
```

```python
def collect_pentas_gaps(penta):
    overall = 0
    p = list(penta)
    for tag in all_tags:
        new_p = tuple([p[0], p[1], tag, p[3], p[4]])
        new_p_count = cnt_pentas[new_p]
        overall += new_p_count
    return overall
```

```python
def collect_tris_gaps_1(tri):
    overall = 0
    p = list(tri)[0:3]
    for tag in all_tags:
        new_p = tuple([p[0], p[1], tag])
        new_p_count = cnt_tris[new_p]
        overall += new_p_count
    return overall

def collect_tris_gaps_2(tri):
    overall = 0
    p = list(tri)[1:4]
    for tag in all_tags:
        new_p = tuple([p[0], tag, p[2]])
        new_p_count = cnt_tris[new_p]
        overall += new_p_count
    return overall

def collect_tris_gaps_3(tri):
    overall = 0
    p = list(tri)[2:5]
    for tag in all_tags:
        new_p = tuple([tag, p[1], p[2]])
        new_p_count = cnt_tris[new_p]
        overall += new_p_count
    return overall
```
