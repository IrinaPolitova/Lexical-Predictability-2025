# Получить словарь

```python
with open('all_verbs_dict.txt', encoding = 'utf-8') as file:
    sem_dict = file.readlines()
```

```python
sem = []
for s in sem_dict:
    if len(s) > 6:
        sem.append(s)

sem_dict = []
for s in sem:
    if s.endswith('\n'):
        s = s.strip('\n')
        sem_dict.append(s)
```

```python
len(sem_dict)
```

```python
import re

pattern_word = '[а-яА-ЯёЁ]*'
pattern_field = '[0-9.]+'
```

```python
sem_dict_dict = []

for i in range(len(sem_dict)):
    s = sem_dict[i]
    
    new_word = re.findall(pattern_word, s)[0]
    new_field = re.findall(pattern_field, s)[0]

    new_list = (new_word.lower(), new_field)

    sem_dict_dict.append(new_list)
```

```python
sem_dict_dict = list(set(sem_dict_dict))
```

```python
sem_dict = []

for i in range(len(sem_dict_dict)):
    new_dict = {}
    new_pair = sem_dict_dict[i]
    
    new_word = new_pair[0]
    new_field = new_pair[1]

    new_dict['word'] = new_word
    new_dict['field'] = new_field

    sem_dict.append(new_dict)
```

```python
import pandas as pd
df = pd.DataFrame(sem_dict)
df.to_csv('sem_dict.csv', index=False)
```

# Узнать объем поля

```python
import pandas as pd
sem_dict = pd.read_csv('sem_dict.csv')
my_verbs = pd.read_csv('POSes_verbs_only.csv')
```

```python
from collections import Counter
cnt = Counter()
```

```python
for tag in list(sem_dict['field']):
    cnt[tag] += 1
```

```python
cnt
```

```python
for index, row in my_verbs.iterrows():
    new_tag = row['sem_tag']

    sum_of_cnt = 0
    for k in cnt.keys():
        if k.startswith(new_tag):
            sum_of_cnt += cnt[k]

    
    print(sum_of_cnt)
```
